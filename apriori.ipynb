```python
#Fonction pour supprimer les redondances dans une liste (elle pourra etre utile)
def suppRedondances(lst):
    lstSansRed=list()
    for i in lst:
        if i not in lstSansRed:
            lstSansRed.append(i)
    return lstSansRed
suppRedondances([1,2,3,2,1,3,4,5,4,3,6,1,1])
```

```python
def creationNvxDataset (dataframe=df):
    dictionnaire=dict()
    for i in dataframe.values:
        l=list()
        dictionnaire[i[0]]=l
    for i in dataframe.values:
        l=dictionnaire[i[0]]
        l.append(i[1])
        l=suppRedondances(l)
        dictionnaire[i[0]]=l
    dataf=pd.DataFrame({'Itemset':dictionnaire.values()})
    dataf.index=dictionnaire.keys()
    return dataf
NewDataFrame = creationNvxDataset()
# NewDataFrame
```

```python
nombreTransaction=len(NewDataFrame.index.tolist())
nombreTransaction
```

```python
def frequence(a,dataframe=NewDataFrame): 
    cpt=0
    for i in dataframe.values:
        i=i.tolist()
        i=i[0]
        for j in i:
            if a==j :
                cpt=cpt+1
    return cpt
print(frequence(24.0))
```

```python
def Items1():
    listvaleurs=list()
    for i in NewDataFrame.values:
        i=i.tolist()
        i=i[0]
        for j in i:
            if(j not in listvaleurs):
                listvaleurs.append(j)
    return listvaleurs
Items=Items1()
Items
```

```python
def Apriori(minsup,minconf):
    #C1
    C1=frequence1()
    L1= dict((c,v) for c,v in C1.items() if v>=minsup) #L1
    print("L1",L1)
    print("\n")
    C2=dict((c,v) for c,v in Frequence2().items() if v>=minsup)
    L2=dict()
    for c,v in C2.items():
        if(c[0] in L1.keys() and c[1] in L1.keys()):
            L2[c]=v
    print("L2",L2)
    print("\n")
    C3=dict((c,v) for c,v in Frequence3().items()if v>=minsup)
    #print("C3",C3)
    #print(L2.keys())
    L3=dict()
    for c,v in C3.items():
        for i in L2.keys():
            for j in L2.keys():
                for k in L2.keys():
                    if(i[0]==j[0] and i[1]!=j[1] and k[0]==i[1] and k[1]==j[1]):
                        if(c[0]==i[0] and c[1]==i[1] and c[2]==j[1]):
                            L3[c]=v
    print("L3",L3)
    print("\n")
    C4=dict((c,v) for c,v in Frequence4().items() if v>=minsup)
    L4=dict()
    for c,v in C4.items():
        for i in L3.keys():
            for j in L3.keys():
                if(i[0]==j[0] and i[1]==j[1]):
                    if(c[0]==i[0] and c[1]==i[1] and c[2]==i[2] and c[3]==j[2]):
                        L4[c]=v
    print("L4",L4)
    print("\n")
    C5=dict((c,v) for c,v in Frequence5().items() if v>=minsup)
    L5=dict()
    for c,v in C5.items():
        for i in L4.keys():
            for j in L4.keys():
                if(i[0]==j[0] and i[1]==j[1] and i[2]==j[2]):
                    if(c[0]==i[0] and c[1]==i[1] and c[2]==i[2] and c[3]==i[3] and c[4]==j[3]):
                        L5[c]=v
    print("L5",L5)
    print("\n")
    C6=dict((c,v) for c,v in Frequence6().items() if v>=minsup)
    L6=dict()
    for c,v in C6.items():
        for i in L5.keys():
            for j in L5.keys():
                if(i[0]==j[0] and i[1]==j[1] and i[2]==j[2] and i[3]==j[3]):
                    if(c[0]==i[0] and c[1]==i[1] and c[2]==i[2] and c[3]==i[3] and c[4]==i[4] and c[5]==j[4]):
                        L6[c]=v
    print("L6",L6)
    print("\n")
    L=L1|L2|L3|L4|L5|L6
    print("L",L)
    minsupAssoc=minsup/39
    ReglesAssociations2=dict()
    A2=RegleCorrélation2(items=L2)
    Rg=dict()
    for c,v in A2.items():
        if(v[0]>=minsupAssoc and v[1]>=minconf):
            ReglesAssociations2[c]=v
            Rg[c]=v
    print("\n\n\nRegles d'association avec 1 conséquent:",ReglesAssociations2)
    
    ReglesAssociations3=dict()
    A3=RegleCorrélation3(items=L3)
    for c,v in A3.items():
        if(v[0]>=minsupAssoc and v[1]>=minconf):
            ReglesAssociations3[c]=v
            Rg[c]=v
    print("\n\n\nRegles d'association avec 2 conséquents:",ReglesAssociations3)
    
    ReglesAssociations4=dict()
    A4=RegleCorrélation4(items=L4)
    for c,v in A4.items():
        if(v[0]>=minsupAssoc and v[1]>=minconf):
            ReglesAssociations4[c]=v
            Rg[c]=v
    print("\n\n\nRegles d'association avec 3 conséquents:",ReglesAssociations4)
    
    ReglesAssociations5=dict()
    A5=RegleCorrélation5(items=L5)
    for c,v in A5.items():
        if(v[0]>=minsupAssoc and v[1]>=minconf):
            ReglesAssociations5[c]=v
            Rg[c]=v
    print("\n\n\nRegles d'association avec 4 conséquents:",ReglesAssociations5)
    
    ReglesAssociations6=dict()
    A6=RegleCorrélation6(items=L6)
    for c,v in A6.items():
        if(v[0]>=minsupAssoc and v[1]>=minconf):
            ReglesAssociations6[c]=v
            Rg[c]=v
    print("\n\n\nRegles d'association avec 5 conséquents:",ReglesAssociations6)
    
    Rg= dict((c,"("+str(v[0])+" , "+str(v[1])+" , "+str(v[2])+")") for c,v in Rg.items())
    dataf=pd.DataFrame({'Les motifs frequents':L.keys(),'Occurence':L.values()})
    dataf2=pd.DataFrame({'Regles d\'association':Rg.keys(),'( Support, Confiance, Lift )':Rg.values()})
    return dataf,dataf2
```

```python
def Recommendations(cat):
    ReglesAssoc=RegleAssociation2()
    ReglesAssoc.update(RegleAssociation3())
    ReglesAssoc.update(RegleAssociation4())
    ReglesAssoc.update(RegleAssociation5())
    ReglesAssoc.update(RegleAssociation6())
    MatchingRules=dict()
    for k in ReglesAssoc.keys():
        l=k.split()
        predecesseurs=list()
        for e in l:
            if e=='->':
                break
            else:
                predecesseurs.append(e[:4])
        predecesseurs=set(predecesseurs)
        if cat==predecesseurs:
            MatchingRules[k]=ReglesAssoc[k]
#     print(MatchingRules)
    listK=list(MatchingRules.keys())
    listK=tri_pas_confidance(listK, MatchingRules)
    lstConsequents=dict()
    for e in listK:
        l=e.split()
        consequents=list()
        b=False
        for c in l:
            if b==True:
                consequents.append(c[:4])
            if c=='->':
                b=True
        lstConsequents[tuple(consequents)]=MatchingRules[e][1]
    print()
    return pd.DataFrame({'Recommendation':lstConsequents.keys(),'Confiance':lstConsequents.values()})
    
def tri_pas_confidance(lst, dictionnaire):
    for i in range(1,len(lst)):
        en_cours = lst[i]
        j = i
        #décalage des éléments du de la liste
        while j>0 and dictionnaire[lst[j-1]][1]<dictionnaire[en_cours][1]:
            lst[j]=lst[j-1]
            j = j-1
        #on insère l'élément à sa place
        lst[j]=en_cours 
    return lst
```
